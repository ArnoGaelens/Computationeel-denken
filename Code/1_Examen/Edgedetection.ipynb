```python
# importeer libraries
from PIL import Image
import math 
import numpy as np

print("optie 11: edge detection")

# converteer een ingelezen afbeelding naar drie matrices
input_image = Image.open("./beeld/landschap_orig.jpg")
r_image_in, g_image_in, b_image_in = input_image.split()
r_in = np.uint32(np.array(r_image_in))
g_in = np.uint32(np.array(g_image_in))
b_in = np.uint32(np.array(b_image_in))

# hier komen de matrixbewerkingen
r_out = np.zeros((r_in.shape[0], r_in.shape[1]))
g_out = np.zeros((g_in.shape[0], g_in.shape[1]))
b_out = np.zeros((b_in.shape[0], g_in.shape[1]))
        
for i in range(1,r_in.shape[0]-1,1):
    print("Processing row ",i," of ",r_in.shape[0]-2)
    for j in range(1,r_in.shape[1]-1,1):      

        r0, g0, b0 = r_in[i-1][j-1],g_in[i-1][j-1],b_in[i-1][j-1]
        r1, g1, b1 = r_in[i-1][j],g_in[i-1][j],b_in[i-1][j]
        r2, g2, b2 = r_in[i-1][j+1],g_in[i-1][j+1],b_in[i-1][j+1]
        
        r3, g3, b3 = r_in[i][j-1],g_in[i][j-1],b_in[i][j-1]
        r4, g4, b4 = r_in[i][j],g_in[i][j],b_in[i][j]
        r5, g5, b5 = r_in[i][j+1],g_in[i][j+1],b_in[i][j+1]
        
        r6, g6, b6 = r_in[i+1][j-1],g_in[i+1][j-1],b_in[i+1][j-1]
        r7, g7, b7 = r_in[i+1][j],g_in[i+1][j],b_in[i+1][j]
        r8, g8, b8 = r_in[i+1][j+1],g_in[i+1][j+1],b_in[i+1][j+1]
        
        avgrv = -1*r0 + 0*r1 + 1*r1 -2*r3 +0*r4 +2*r5 -1*r6 + 0*r7 +1*r8
        avggv = -1*g0 + 0*g1 + 1*g1 -2*g3 +0*g4 +2*g5 -1*g6 + 0*g7 +1*g8
        avgbv = -1*b0 + 0*b1 + 1*b1 -2*b3 +0*b4 +2*b5 -1*b6 + 0*b7 +1*b8
        
        avgrv=avgrv/9
        avggv=avggv/9
        avgbv=avgbv/9
        
        avgrh = -1*r0  -2*r1 -1*r1 +0*r3 +0*r4 +0*r5 +1*r6 + 2*r7 +1*r8
        avggh = -1*g0  -2*g1 -1*g1 +0*g3 +0*g4 +0*g5 +1*g6 + 2*g7 +1*g8
        avgbh = -1*b0  -2*b1 -1*b1 +0*b3 +0*b4 +0*b5 +1*b6 + 2*b7 +1*b8
                
        avgrh=avgrh/9
        avggh=avggh/9
        avgbh=avgbh/9
        
        newr = math.sqrt(avgrv**2 + avgrh**2)
        newg = math.sqrt(avggv**2 + avggh**2)
        newb = math.sqrt(avgbv**2 + avgbh**2)
        
        
        #to black and white
        threshold = 5
        avg = (newr+newg+newb)/3
        if avg < threshold:
            r_out[i][j] = 0
            g_out[i][j] = 0
            b_out[i][j] = 0
        else:
            r_out[i][j] = 255
            g_out[i][j] = 255
            b_out[i][j] = 255

# converteer drie matrices terug naar een afbeelding
r_image_out = Image.fromarray(np.uint8(r_out))
g_image_out = Image.fromarray(np.uint8(g_out))
b_image_out = Image.fromarray(np.uint8(b_out))
output_im = Image.merge("RGB", (r_image_out, g_image_out, b_image_out))
output_im.save("./beeld/landschap_edge.jpg")
output_im.show()
```
