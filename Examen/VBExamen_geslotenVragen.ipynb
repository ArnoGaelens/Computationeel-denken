### 2.2 Welke waarden worden er geprint op regel 12?
    Tabel maken: 15 rijen. Lijst b in kolommen. Dan telkens modulo doorlopen.
    [True, False, False, False]
    
    

````python
lijstA = [1,3,5,6]
lijstB = [True,False,True,False]
count = 0
    
for i in range(len(lijstA)): #LET OP: range(4) betekent: 0 tot 4(zonder 4)
    c = lijstA[i]
    for j in range(c): # Range(1) betekent: 0 tot 1 (zonder 1) (Zie voorbeeelden hieronder)
        p = j%len(lijstB)   #p = modulo van j / 4  (0%x = 0)
        lijstB[p] = not lijstB[p] # gewoon not(True) = False
        print(lijstB) 
        count = count + 1 # Globale counter dus bijhouden tot einde
        

print(count)
print(lijstB)
````

### 3.1 Wat is de waarde die geprint wordt op regel 14? 
-2
### 3.2 Welke waarden worden er geprint op regel 15?
7
### 3.3 Welke waarden worden er geprint op regel 16?
49,5

````python
lijstX = [3.5,7.8,6.9,2.1,0.7,9.9]
result = [0,0,0]

for e in lijstX:
    e = int(e*10)
    if e < 10:
        result[0]+=e
    elif e<50:
        result[1] -= 1
    else:
        result[2] = e/2
t = result[1]
result[1] = result[0]
result[0] = t
print(result[0])
print(result[1])
print(result[2])
````
