## Opgave:
    Implementeer een algoritme dat op basis van een oorspronkelijke figuur en een patroon bepaalt of het patroon voorkomt in de figuur. Het algoritme dat wordt uitgevoerd heeft dus een waarde True of False terug.

    Om na te gaan of een patroon in een figuur voorkomt wordt de kleur van elk pixel (i.e. elke positie) in de oorspronkelijke figuur vergeleken met de kleurwaarde van de pixel in de linkerbovenhoek van het patroon. Indien deze waarden overeenkomen, dan wordt drie keer een hulpmethode matchingPatternTest opgeroepen. Telkens wordt hierbij voor een bepaalde kleurwaarde (rood, groen en blauw) nagegaan of de waarden van alle pixels in de oorspronkelijke figuur vanaf die bepaalde positie overeenkomen met het patroon. Indien dat zo is, dan geeft de methode True terug, anders False. Indien de drie testen slagen, dan komt het patroon immers voor in de oorspronkelijke figuur.

```python
# importeer libraries
import math 
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


# converteer een ingelezen afbeelding naar drie matrices
input_image = Image.open("./beeld/landschap1.jpg")
r_image_in, g_image_in, b_image_in = input_image.split()
r_in = np.uint32(np.array(r_image_in))
g_in = np.uint32(np.array(g_image_in))
b_in = np.uint32(np.array(b_image_in))

pattern_image = Image.open("./beeld/landschap1 - Copy.jpg")
r_image_pat, g_image_pat, b_image_pat = pattern_image.split()
r_pat = np.uint32(np.array(r_image_pat))
g_pat = np.uint32(np.array(g_image_pat))
b_pat = np.uint32(np.array(b_image_pat))

#-----------------------------------------------------------------------------------------
# hier komen de matrixbewerkingen
def matchingPatterntest(waardenFiguur,  waardePatroon, i, j):
    for e in range(waardePatroon.shape[0]):
        for y in range(waardePatroon.shape[1]):
            if waardePatroon[e][y] != waardenFiguur[i+e][j+y]:
               return False
        return True

match_found = False

for i in range(r_in.shape[0]):
    for j in range(r_in.shape[1]):
        if r_in [i][j] == r_pat[0][0] and g_in [i][j] == g_pat[0][0] and b_in [i][j] == b_pat[0][0]:
            if (matchingPatterntest(r_in, r_pat, i, j) and 
            matchingPatterntest(g_in, g_pat, i, j) and 
            matchingPatterntest(b_in, b_pat, i, j)):
                match_found = True
                break
    if match_found:
        break

if match_found:
    print(True)
else:
    print(False)



# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(np.dstack((r_in, g_in, b_in)))
plt.title('Original Image')

# Display the pattern image
plt.subplot(1, 2, 2)
plt.imshow(np.dstack((r_pat, g_pat, b_pat)))
plt.title('Pattern Image')

# Show the plots
plt.show()
  
```

# Deel C: tribonacci
Tribonacci: Schrijf een algoritme om het n-de getal in de rij van Tribonacci te vinden. Dit is een variant op de rij van Fibonacci waar de eerste drie getallen 0, 1, 1 zijn, en alle volgende getallen de som van de vorige drie.
Het begin van de rij ziet er uit als volgt:

0,1,1,2,4,7,13,24,44,81,149, 

Schrijf het volledige Python algoritme om het n-de getal in deze rij te vinden. Dit algoritme moet werken voor ieder natuurlijk getal n>3.

```python
import numpy
def tribonacci(n):
    lijst = []
    lijst[0] = 0
    lijst[1] = 1
    lijst[2] = 1
    for i in range(3,n):
        getali = lijst[i-1] + lijst[i-2] + lijst[i-3]
        lijst.append(getali) 
    return lijst




```
