# Oefening 1a

### Gemiddelde van een array-reeks van 5 getallen

Lees 5 gehele getallen in en stop ze in een array.  

Doorloop de array door middel van de index-notatie [ ] en laat deze getallen één voor één zien op 1 lijn op het scherm. 

Bepaal vervolgens het rekenkundig gemiddelde. Maak geen gebruik van de ingebouwde functie np.mean(...).  

Probeer ten slotte ook eens de ingebouwde functie np.mean(...) uit.  

Definiëer 3 eigen functies: **inlezen_rij(), uitprinten_rij(...) en bepaal_gemiddelde(...)** die allen dezelfde array als argument hanteren.  

Het programma kan er ongeveer als volgt uitzien:  

>    definieer een array met de naam "mijn_rij"
>    
>    mijn_rij = inlezen_rij()
>    
>    uitprinten_rij(mijn_rij)
>    
>    gemiddelde = bepaal_gemiddelde(mijn_rij)
>    
>    toon het gemiddelde op het scherm  
    
##### Oplossing

`````python
# Met arrays

import numpy as np


def inlezen_rij():
    mijn_rij = np.zeros((5), dtype=int)
    for j in range(mijn_rij.shape[0]):
        getal = int(input("Geef een getal in: "))
        mijn_rij [j] = getal
    return mijn_rij


def gemiddelde(reeks):
    som = 0
    for j in range(reeks.shape[0]):
        som += reeks[j]     #Let op voor deze. niet gewoon j gebruiken
    gemiddelde = som/reeks.shape[0]
    return gemiddelde

def gemiddeldenp(reeks):
    gemiddeldenp = np.mean(reeks)
    return gemiddeldenp

print(inlezen_rij())
print(f'Het gemiddelde adhv eigen berekening is: {gemiddelde(mijn_rij)}')
print(f'Het gemiddelde adhv np.mean(): {gemiddeldenp(mijn_rij)}')


`````

# Oefening 1b

### Gemiddelde van een array-reeks van een willekeurig aantal getallen

Pas uw vorige oplossing aan zodat ze altijd werkt, en niet enkel voor precies 5 getallen.
Het aantal getallen is op voorhand niet gekend.  

Je geeft getallen in en stopt met ````<enter>````, zonder meer.  

De functie **inlezen_rij()** kan er ongeveer zo uit zien:  
> declareer een lege **list**  

>````python
> while True:
>    getal_string = input("Geef waarde: ")
>    if getal_string == "":
>        break
>    getal = int(getal_string)
>    lijst.append(getal)
>````    

> maak aan array op basis van de lijst
  
>    lever de array af (return)  
##### Oplossing
    

`````python
import numpy as np


def inlezen_rij():
    lijst = []
    while True:
        invoer = input("Geef een getal in om aan de rij toe te voegen:")
        if invoer == "":
            break
        else:
            getal = int(invoer)
            lijst.append(getal)
        print(lijst)
    mijn_rij = np.zeros((len(lijst)), dtype=int)

    for j in range(mijn_rij.shape[0]):
        mijn_rij [j] = lijst[j]
    return mijn_rij

def gemiddelde(reeks):
    som = 0
    for j in range(reeks.shape[0]):
        som += reeks[j]     #Let op voor deze. niet gewoon j gebruiken
    gemiddelde = som/reeks.shape[0]
    return gemiddelde

def gemiddeldenp(reeks):
    gemiddeldenp = np.mean(reeks)
    return gemiddeldenp

mijn_rij = inlezen_rij()

print(mijn_rij)
print(f'Het gemiddelde adhv eigen berekening is: {gemiddelde(mijn_rij)}')
print(f'Het gemiddelde adhv np.mean(): {gemiddeldenp(mijn_rij)}')
`````

# Oefening 2

### Samenvoegen van 2 arrays

Lees twee arrays in. Voeg de waarden van beide arrays samen in een derde array.  
Print deze samengevoegde array uit.  

##### Voorbeeld

> rij1: 5 8 3 11  
> rij2: 4 7 14 21 10   
> rij3 wordt: 5 8 3 11 4 7 14 21 10  

Doe het samenvoegen zelf, element per element, door gebruik te maken van de index-notatie [ ]. Maak met andere woorden geen gebruik van de functie np.hstack(...).  

Declareer de nodige functies.

Het programma kan er ongeveer uitzien als volgt:

```` python
rij1 = inlezen_rij()
rij2 = inlezen_rij()
rij3 = samenvoegen_rij(rij1, rij2)

Toon de rijen
````

Probeer uiteindelijk facultatief ook even de functie np.hstack(...) uit.  

##### Oplossing


`````python
import numpy as np



def inlezen_rij():
    lijst = []
    while True:
        invoer = input("Geef een getal in om aan de rij toe te voegen:")
        if invoer == "":
            break
        else:
            getal = int(invoer)
            lijst.append(getal)
        print(lijst)
    mijn_rij = np.zeros((len(lijst)), dtype=int)

    for j in range(mijn_rij.shape[0]):
        mijn_rij [j] = lijst[j]
    return mijn_rij

def samenvoegen(een, twee):
    samen = np.zeros((een.shape[0]+twee.shape[0]), dtype=int)
    for i in range(samen.shape[0]):
        if i < een.shape[0]:
            samen[i] = een[i]
        else:
            samen[i] = twee[i-een.shape[0]]
    return samen

            
rij1 = inlezen_rij()
rij2 = inlezen_rij()
rij3 = samenvoegen(rij1, rij2)

print(f'Samengevoegd vormen de 2 rijen de volgende rij{rij3}')



    
`````

# Oefening 3

### Reeks van Fibonacci

Genereer n getallen uit de reeks van Fibonacci, waarbij n ingegeven wordt door de gebruiker.  
Stop deze n getallen in een array.  

De reeks van Fibonacci is als volgt opgebouwd:  

> ````f1 = 1   f2 = 1   f3 = 2   f4 = 3   f5 = 5    . . . ````   
> 
> fi = fi-1 + fi-2  voor  i>2

Vanaf de derde waarde is elk getal de som van de vorige twee.  

Schrijf vervolgens de rij uit op het scherm.  

##### Oplossing

`````python
import numpy as np

def fibonacci(n):
    fn = np.zeros(n, dtype=int)
    fn[0] = 1
    fn[1] = 1
    for i in range(2,fn.shape[0]):
        fn[i] = fn[i-1]+fn[i-2]
    return fn

aantal = int(input("Gewenste aantal Fibonacci: "))
fn = fibonacci(aantal)
print(fn)







`````

> 8    9    2    5  
> 
> De grootste waarde is 9 en deze waarde komt 3 maal voor op posities:

> rij: 1 kolom: 0  
> rij: 2 kolom: 1  

Declareer de nodige functies.

Het programma kan er ongeveer uitzien als volgt:

````python
m = inlezen_matrix()
uitprinten_matrix(m)
grootste = bepaal_grootste(m)
aantal_max = bepaal_frequentie(m, grootste)

toon de waarde van het maximum evenals de frequentie

toon_posities(m, grootste)
````
##### Oplossing


`````python
import numpy as np


def inlezen_matrix(rijen,kolommen):
    matrix = np.zeros((rijen,kolommen),dtype=float)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            matrix [i][j] = int(input(f'Element [{i}][{j}]:'))
    return matrix

def max_matrix(matrix):
    max = 0
    teller = 0
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            if matrix[i][j] > max:
                max = matrix[i][j]
                teller = 1
            elif matrix [i][j] == max:
                teller += 1    
    return max, teller

tabel = inlezen_matrix(int(input("Aantal rijen: ")),int(input("Aantal kolommen: ")))
maxwaarde, aantal_max = max_matrix(tabel)

print(f'De tabel ziet er als volgt uit: \n {tabel}')
print(f'De maximumwaarde bedraagt {maxwaarde} en komt {aantal_max} keer in te tabel voor')







`````
