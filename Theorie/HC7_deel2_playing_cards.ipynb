```python
from random import randint

rank = ("2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King", "Ace")
suit = ("Clubs", "Diamonds", "Hearts", "Spades")

RANKS = len(rank)
SUITS = len(suit)
N = RANKS * SUITS 

# make a new deck of cards
def deck_of_cards():
    deck = []
    for r in rank:
        for s in suit:
            card = (r, s)
            deck.append(card)    
    return deck

# shuffle the deck of cards 
def shuffle_deck(deck):    
    for i in range (N):
        r = randint(i, N-1)
        temp = deck[r]
        deck[r] = deck[i]
        deck[i] = temp
    return deck

# cut deck at given position
def cut_deck(deck, position):
    deck = deck[0:position]
    return deck

# split in two decks - return second deck
def split_deck(deck):
    deck_1 = []
    deck_2 = []
    counter = 0
    for card in deck:
        if (counter % 2 == 0):
            deck_1.append(card)            
        else: 
            deck_2.append(card)
        counter += 1
    return deck_2

# execute splitting four times
def iterate_splitting(deck):
    for i in range(4):
        deck = split_deck(deck)
    return deck

# preparing the trick
deck = deck_of_cards()
shuffled_deck = shuffle_deck(deck)
assert len(shuffled_deck) == 52, "er zijn geen 52 kaarten"
target_card = deck[15]

# executing the trick
split_position = randint(16, 31) # generate random getal 16 <= getal <= 31
assert (split_position >= 16 and split_position <= 31)
deck_remainder = cut_deck(shuffled_deck, split_position)

# showing the magic
remaining_card = iterate_splitting(deck_remainder)
print (remaining_card[0] == target_card)
    
    
```
